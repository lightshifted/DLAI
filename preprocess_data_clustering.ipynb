#### Categorization

```python
# store unique urls in pandas DataFrame
m = pd.DataFrame([])
m['url'] = pageviews['Last Page Seen Current Value'].unique()

# list of indices for sample tracking
idx = m.index.to_list()

# list of column names
df_columns = ['scheme', 'netloc', 'path'
              , 'params', 'query', 'fragment']

# null dataframe with columns df_columns and index idx
df = pd.DataFrame([], columns=df_columns, index=idx).fillna(0)

# segment urls for categorization
df['scheme'] = [urlparse(m['url'][i]).scheme for i in range(len(idx))]
df['netloc'] = [urlparse(m['url'][i]).netloc for i in range(len(idx))]
df['path'] = [urlparse(m['url'][i]).path for i in range(len(idx))]
df['params'] = [urlparse(m['url'][i]).params for i in range(len(idx))]
df['query'] = [urlparse(m['url'][i]).query for i in range(len(idx))]
df['fragment'] = [urlparse(m['url'][i]).fragment for i in range(len(idx))]

# show df
df.head(5)
```
